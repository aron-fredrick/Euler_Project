```python
import time
```

```python
start = time.time()
nmax = 100

sumList = [[1,1]]

for ii in range(2,nmax):
    start1 = time.time()
    prev = len(sumList)
    sumList.append([ii])
    sortedLists = []
    for listA in sumList:
        listA.append(1)
        testA = listA[0:-2]+[listA[-2]+1]
        #print(listA,testA,listA[-2]+1)
        if len(testA)>1 and testA[-1]<=testA[-2]:
            sortedLists.append(testA)
        
    sumList = sumList + sortedLists
    if ii%1 == 0:
        print(ii,len(sumList),2*prev+1 - len(sumList),time.time()-start1)
    
end = time.time()
print(end-start)
print(len(sumList))
```

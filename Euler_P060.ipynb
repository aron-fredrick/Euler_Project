```python
import numpy as np

def isprime(n):
    # checks to see if a number, n, is prime
    import numpy as np
    prime = True
    if n>=2:
        
        for ii in range(2,int(np.ceil(np.sqrt(n)))):
            if n%ii == 0:
                prime = False
                break
    else:
        prime = False
    return prime

def get_primes(n):
    # same as primes, but faster
    m = n+1
    numbers = [True] * m
    for i in range(2, int(n**0.5 + 1)):
        if numbers[i]:
            for j in range(i*i, m, i):
                numbers[j] = False
    primes1 = []
    for i in range(2, m):
        if numbers[i]:
            primes1.append(i)
    return primes1

def concat_check(nlist):
    # check if all the various concatenations in a list come out prime
    all_prime = True
    index = 0
    while index<len(nlist) and all_prime:
        
        n1 = nlist[index]
        nlist2 = nlist[0:index]+nlist[index+1:]
        
        for n2 in nlist2:
            ntest = int(str(n1)+str(n2))
            if not isprime(ntest):
                all_prime = False
                break
        
        index += 1
    
    return all_prime
    
    
def next_indices(pindices):
    # gives the next set of indices after a given set, increasing 1 index by one
    pdiff = list(np.diff(pindices))
    
    #inx = len(pdiff)-1-pdiff[::-1].index(1)
    inx = 0
    while inx < len(pdiff) and pdiff[inx] == 1:
        inx+=1
    
    return list(range(0,inx))+[pindices[inx]+1]+pindices[inx+1::]    
```

```python
plist = get_primes(10**4)
plist.remove(2)
plist.remove(5)
```

```python
groups = []

for prime in plist:
    
    cond = False
    
    if prime == 2:
        continue
        
    plays_well = False
    for group in groups:
        test_group = group+[prime]
        if concat_check(test_group):
            group.append(prime)
            plays_well = True
            if len(group) == 4:
                cond = True
                
            
    if not plays_well:
        groups = groups + [[prime]]
    
    if cond:
        break

groups
```

```python
#returns subgroups where each element concatenates with the first element of subgroup
groups = []

for ii in range(0,len(plist)):

    prime = plist[ii]
    concat = [prime]

        
    plist2 = plist[:ii]+plist[ii+1:]
    for prime2 in plist2:
        if concat_check([prime,prime2]):
            concat.append(prime2)
            
    groups.append(concat)
```

```python
# returns subgroups where every element concats with the first 2 (makes the groups really small)
groups2 = []
for jj in range(0,len(groups)):

    test_group = groups[jj][1:]
    prime1 = groups[jj][0]


    for ii in range(0,len(test_group)):

        prime = test_group[ii]
        concat = [prime]


        plist2 = test_group[:ii]+test_group[ii+1:]
        for prime2 in plist2:
            if concat_check([prime,prime2]):
                concat.append(prime2)

        groups2.append([prime1]+concat)
                
```

```python
psum = 10**12
pindices = [0,1,2,3,4]
nprimes = 5
pgroup = []

for group in groups2:
    #print(group[0])
    pindices = [0,1,2,3,4]
    group.sort()
    if (len(group)>(nprimes-1)) and (group[nprimes-1] < psum):
        while (pindices[-1] < len(group)) and (group[pindices[-1]] < psum):

            #print(pindices)
            
            test_primes = [group[ii] for ii in pindices]
            test_sum = sum(test_primes)

            if test_sum < psum:
                cond = concat_check(test_primes)

                if cond: 
                    psum = test_sum
                    pgroup = test_primes

            pindices = next_indices(pindices)
```
